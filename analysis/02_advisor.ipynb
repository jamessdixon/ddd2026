```python
from azure.identity import DefaultAzureCredential
from azure.mgmt.advisor import AdvisorManagementClient
import pandas as pd
```

```python
# DefaultAzureCredential tries multiple auth methods in order:
# 1. Environment variables, 2. Managed Identity, 3. Azure CLI, 4. VS Code, etc.
credential = DefaultAzureCredential()
```

```python
credential = DefaultAzureCredential()
SUBSCRIPTION_ID = "3ca6ed87-dfc5-4dce-9cf7-71c84695304d"

advisor_client = AdvisorManagementClient(credential, SUBSCRIPTION_ID)

recommendations = list(advisor_client.recommendations.list())

rows = []
for r in recommendations:
    rows.append({
        "Category":      r.category,
        "Impact":        r.impact,
        "ImpactedArea":  r.impacted_field,
        "ImpactedValue": r.impacted_value,
        "Description":   r.short_description.problem if r.short_description else None,
        "Solution":      r.short_description.solution if r.short_description else None,
        "ResourceGroup": r.resource_group if hasattr(r, "resource_group") else None,
    })

df = pd.DataFrame(rows)
display(df)
```

```python
df.to_csv("advisor_recommendations.csv", index=False)
```
